```python
%matplotlib inline
import math
import time
import numpy as np
from sklearn.linear_model import LinearRegression
import  statsmodels.api as sm 
import torch
from d2l import torch as d2l
```

## Generate Synthetic data 

```python
# help(d2l.SyntheticRegressionData)
```

```python
synthetic_data_weights = torch.tensor([2, -3.4, 1, -8, 4, 3, -9, 1,  2, -6])
synthetic_data_b = 4.2
num_dim =  synthetic_data_weights.shape[0]
```

```python
num_dim
```

```python
data = d2l.SyntheticRegressionData(w=synthetic_data_weights, 
                                   b=synthetic_data_b,
                                  num_train=1000,
                                  num_val=1000,
                                  batch_size=100,
                                  noise=0.5)
```

```python
# data.X = data.X[:,:8]
# num_dim =  data.X.shape[1]
```

```python
data.y.shape
```

## Linear regression using sklearn 

```python
nX = data.X.detach().numpy()
ny = data.y.detach().numpy()
```

```python
sk_model = LinearRegression()
sk_model.fit(nX, ny)
```

```python
sk_model.coef_, sk_model.intercept_
```

## Linear regression using statsmodels

```python
sm_mdl = sm.OLS(endog=ny, exog=sm.add_constant(nX)).fit()
```

```python
sm_mdl.params
```

## Linear Regression from scratch 

```python
model_scratch = d2l.LinearRegressionScratch(num_inputs=num_dim, lr=0.01, sigma=1000)
trainer = d2l.Trainer(max_epochs=100)
trainer.fit(model_scratch, data)
```

```python
model_scratch.w, model_scratch.b
```

## Linear Regression based on pytorch

```python
model = d2l.LinearRegression(lr=0.01)
trainer = d2l.Trainer(max_epochs=100)
trainer.fit(model, data)
```

```python
model.get_w_b()
```

```python
LL = [int(f) for f in list(np.geomspace(5, 10000, 20, endpoint = True)) ]
```

```python
LL
```

```python
for N in LL:
    data = d2l.SyntheticRegressionData(w=synthetic_data_weights, 
                                       b=synthetic_data_b,
                                      num_train=N,
                                      num_val=N,
                                      batch_size=100,
                                      noise=0.5)
    model_scratch = d2l.LinearRegressionScratch(num_inputs=num_dim, lr=0.01, sigma=0.1)
    trainer = d2l.Trainer(max_epochs=3)
    trainer.fit(model_scratch, data)
    
    print(np.sum((data.w - model_scratch.w.reshape(data.w.shape)).abs()) + (data.b - model_scratch.b).abs())
    
    
```
